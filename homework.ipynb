Используйте код из семинара для нахождения кандидатов в дубликаты (шинглы -> минхэш - lsh) и рассчитайте реальную меру Жаккара между полученными кандидатами. Настройте параметры k, num_hash_functions, bands так чтобы результаты получались адекватные (мера Жаккара хотя бы больше нуля).

```python
import hashlib
from collections import defaultdict
from tqdm.auto import tqdm

import codecs

from operator import itemgetter
```

```python
!mkdir data
!wget https://github.com/mannefedov/compling_nlp_hse_course/blob/master/data/anna_karenina.txt -P data
```

```python
file = codecs.open('data/anna_karenina.txt', 'r', 'utf_8_sig')
corpus = file.read()
file.close()
```

```python
text_karenina = corpus.split("\n\n")
```

```python
# генерирует список шинглов из строки:

def get_shingles(text, k=5):
    shingles = set()
    for i in range(len(text) - k + 1):
        shingle = text[i:i + k]
        shingles.add(shingle)
    return shingles
```

```python
# хеширует строку и возвращает число:

def hash_string(s):
    return int(hashlib.md5(s.encode('utf8')).hexdigest(), 16)
```

```python
# генерирует k хеш-функций добавляя индекс к строке:

def generate_hash_functions(k):
    functions = []
    for i in range(k):
        functions.append(lambda x, i=i: hash_string(x + str(i)))
    return functions
```

```python
# вычисляет minhash-сигнатуру для списка шинглов:

def compute_minhash_signature(shingles, hash_funcs):
    if not shingles:  # проверка, если список шинглов пустой
        return None

    signature = []
    for hash_func in hash_funcs:
        min_hash = min(hash_func(shingle) for shingle in shingles)
        signature.append(min_hash)
    return signature
```

```python
# Разрезает сигнатуры на куски (bands), и группирует индексы сигнатур по совпадению кусков:

def lsh(signatures, bands):
    buckets = defaultdict(list)
    band_length = len(signatures[0]) // bands

    for idx, sig in tqdm(enumerate(signatures)):
        if sig is None: # проверка, если сигнатура не была получена, если список шинглов пустой
            continue

        for b in range(0, bands, band_length):
            start = b
            end = start + band_length
            band = tuple(sig[start:end])
            buckets[band].append(idx)
    return buckets
```

```python
# Поиск похожих строк, используя MinHash и LSH:

def find_similar_strings(strings_list, k=5, num_hashes=100, bands=20):
    hash_funcs = generate_hash_functions(num_hashes)
    signatures = []
    shingles_list = []

    # Обработка каждого текста и нахождение шинглов и рассчет сигнатур:
    for string in tqdm(strings_list):
        shingles = get_shingles(string, k)
        if not shingles: # проверка на пустой шингл
            continue

        shingles_list.append(shingles)
        signature = compute_minhash_signature(shingles, hash_funcs)
        if signature is not None:
            signatures.append(signature)

    # Вычисляются кандидаты по кускам сигнатур:
    buckets = lsh(signatures, bands)
    candidates = set()
    for bucket in buckets.values():
        if len(bucket) > 1:
            for i in bucket:
                for j in bucket:
                    if i < j:
                        candidates.add((i, j))

    return candidates
```

```python
# Мера Жаккара:

def get_jaccard(x, y):
    x_sh = get_shingles(x, 10)
    y_sh = get_shingles(y, 10)

    if not x_sh or not y_sh: # Если x_sh или y_sh пустые (=> нет общих элементов)
        return 0.0

    return len(x_sh & y_sh) / len(x_sh | y_sh)
```

```python
# Настраиваемые параметры:

k = 4
num_hashes = 100
bands = 20
```

```python
candidates = find_similar_strings(text_karenina, k=k, num_hashes=num_hashes, bands=bands)
```

```python
# Подсчет меры Жаккара для кандидатов:

similar_pairs = []
for i, j in tqdm(candidates):
    jaccard = get_jaccard(text_karenina[i], text_karenina[j])
    if jaccard > 0:  # Проверка, чтобы сходство кандидатов было больше 0
        similar_pairs.append((i, j, jaccard))
```

```python
similar_pairs
```

```python
similar_pairs = sorted(similar_pairs, key=itemgetter(2), reverse=True)

for i, j, jac in similar_pairs[:10]:
    print(f"Pair ({i}, {j}) Jaccard: {jac:.3f}")
    print(f"Text 1:\n{text_karenina[i][:200]}...\n")
    print(f"Text 2:\n{text_karenina[j][:200]}...\n")
    print("="*50)
```
